```python
df.replace({'male': 0, 'female': 1},
           inplace=True)
df.head()
```

```python
to_remove = ['Weight', 'Duration']
df.drop(to_remove, axis=1, inplace=True)
```

```python
df.head()
```

```python
features = df.drop(['User_ID', 'Calories'], axis=1)
target = df['Calories'].values

X_train, X_val,\
    Y_train, Y_val = train_test_split(features, target,
                                      test_size=0.1,
                                      random_state=22)
X_train.shape, X_val.shape
```

```python
# Normalizing the features for stable and fast training.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()
```

```python
from sklearn.metrics import mean_absolute_error as mae
models = [LinearRegression(), XGBRegressor(),
          Lasso(), RandomForestRegressor(), Ridge()]

for i in range(5):
    models[i].fit(X_train, Y_train)
    print(f'{models[i]} : ')

    train_preds = models[i].predict(X_train)
    print('Training Error : ', mae(Y_train, train_preds))

    val_preds = models[i].predict(X_val)
    print('Validation Error : ', mae(Y_val, val_preds))
    print()
```

```python
# Check the exact feature order from training
print("Features used in training:")
print(features.columns.tolist())

# Select the best model (XGBRegressor had lowest error)
best_model = models[1]  # XGBRegressor

# Make predictions on validation set
val_predictions = best_model.predict(X_val)

# For new data, create with the same feature columns in the same order
new_data = pd.DataFrame({col: [0] for col in features.columns})

# Update with actual values
new_data.loc[0, 'Age'] = 30
new_data.loc[0, 'Height'] = 175
new_data.loc[0, 'Gender'] = 0  # 0 for male, 1 for female
new_data.loc[0, 'Body_Temp'] = 100
new_data.loc[0, 'Heart_Rate'] = 130


# 2. Scale using the same scaler
new_data_scaled = scaler.transform(new_data)

# 3. Predict
predicted_calories = best_model.predict(new_data_scaled)
print(f"Predicted Calories Burnt: {predicted_calories[0]}")
```
